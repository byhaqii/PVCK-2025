```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('/content/drive/MyDrive/PVCK/Images/kitten.jpg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))
plt.imshow(img_rgb)
plt.axis('off')
plt.show()
```

```python
import numpy as np

rg = img_rgb.copy()
rg[:,:,2] = 0  # remove blue channel

gb = img_rgb.copy()
gb[:,:,0] = 0  # remove red channel

plt.subplot(1,2,1)
plt.imshow(rg)
plt.title('Red-Green Only')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(gb)
plt.title('Green-Blue Only')
plt.axis('off')

plt.show()
```

```python
crop_img = img_rgb[10:100, 10:100]
plt.imshow(crop_img)
plt.axis('off')
plt.show()
```

```python
green_channel = img_rgb[5:30, :, 1]
plt.imshow(green_channel, cmap='gray')
plt.axis('off')
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

canvas = np.zeros((400, 400, 3), dtype=np.uint8)

# Box 1 - Red
cv2.rectangle(canvas, (20, 20), (100, 100), (255, 0, 0), -1)

# Box 2 - Green
cv2.rectangle(canvas, (120, 50), (200, 150), (0, 255, 0), -1)

# Box 3 - Blue
cv2.rectangle(canvas, (50, 200), (130, 280), (0, 0, 255), -1)

# Box 4 - Yellow
cv2.rectangle(canvas, (250, 80), (350, 160), (255, 255, 0), -1)

# Box 5 - Purple
cv2.rectangle(canvas, (200, 250), (300, 350), (255, 0, 255), -1)

plt.imshow(cv2.cvtColor(canvas, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
